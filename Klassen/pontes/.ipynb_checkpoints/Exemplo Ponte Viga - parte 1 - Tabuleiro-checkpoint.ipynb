## Esquema estrutural

<img src="pontes/images/1.1.jpg" width="300" />

```python
import pandas as pd
import numpy as np
import pint
u = pint.UnitRegistry()

# distancia entre eixos do trem tipo
a = 2 * u.meter

# distancia entre as longarinas
lx = 6.6 * u.meter

# ????? 
t1 = 0.2 + 2 * (0.085) + 0.25
t2 = 0.5 + 2 * (0.025) + 0.25
t = np.sqrt((t1 + t2)) * u.meter

# Imprimir tabela de dados
df = pd.DataFrame(
    [t1, t2,t, t/a, lx/a],
    ['t1', 't2', 't', 't/a', 'lx/a']
)
df
```

## Cargas permanentes

```python
# Peso próprio da laje
altura_laje = 0.25 * u.meter
peso_concreto = 2.5 * u.ton_force / u.meter**3
pp_laje = altura_laje * peso_concreto

# Peso próprio pavimento
altura_pavimento = 0.085 * u.meter
peso_pavimento =  2.4 * u.ton_force / u.meter**3
pp_pavimento = altura_pavimento * peso_pavimento

# Peso total
g = pp_laje + pp_pavimento

# Imprimir tabela de dados
df = pd.DataFrame(
    [altura_laje.magnitude, peso_concreto, pp_laje, altura_pavimento, peso_pavimento, pp_pavimento, g],
    ['altura_laje', "peso_concreto", 'pp_laje', 'altura_pavimento', 'peso_pavimento', 'pp_pavimento', 'g']
)
df
```

```python
kxm = 0.0417 * u.ton_force
mxm = kxm * g * lx**2

kym = 0.069 * u.ton_force
mym = kym * g * lx**2

kxe = -0.0833 * u.ton_force
mxe = kxe * g * lx**2

# Imprimir tabela de dados
df1 = pd.DataFrame([
    [g.magnitude, g.units],
    [lx.magnitude, lx.units],
        [kxm.magnitude, kxm.units],
    [mxm.magnitude, mxm.units],
        [kym.magnitude, kym.units],
    [mym.magnitude, mym.units],
        [kxe.magnitude, kxe.units],
    [mxe.magnitude, mxe.units]
],
    columns=['magnitude', 'units'],
    index=['g','lx','$k_{xm}$','$M_{xm}$','$k_{ym}$','$M_{ym}$','$k_{xe}$', '$M_{xe}$']
)
df1
```

```python
# definir funcao do momento fletor devido carga movel positiva
lx = lx.magnitude
def ms_movel_neg(x):
    if(x <= 0.35*lx):
        return (mxe_movel*((0.35*lx - x)/(0.35*lx)))
    else: 
        return 0
    
# definir funcao do momento fletor devido carga movel negativa
def ms_movel_pos(x):
    if(x <= 0.30*lx):
        return mxm_movel* ((x)/ (0.3*lx))
    else: return mxm_movel
```

### DImensionamento ao momento fletor

```python
#fck 25
fck = 2.5 # kN/cm2 
fcd = fck/1.35
#Aco CA 50
fyk = 50  #kN/cm2
fs = fyk/1.15
```

```python
# formulas

## KMD
def KMD(md,bw,d,fcd):
    return (md*1000)/(bw * d**2 * fcd)

## Area de Aco
def AS(md,kz,d,fs):
    return ((md)* 1000)/(kz*d*fs)

## Acha o kx
def find_kx(md):
    kx1, kx2 = np.roots([-.0272, 0.68, -md])
    if(kx1 > 0 and kx1< 1):
        return kx1
    else:
        return kx2

## funcao completa
## Acha area de aco a partir do md e geometria    
def md_to_as(md_,bw,d,h,fcd,fs):
    """
    returns: kmd, kx, kz, as_prov, as_min, as_def"""
    md = np.abs(md_)
    kmd_ = KMD(md,bw,d,fcd)
    kx_ = find_kx(kmd_)
    kz_ = 1-(0.4*kx_)
    as_prov = AS(np.abs(md), kz_, d, fs)
    as_def = as_prov
    as_min = 0.0015 * bw * h
    if(as_prov < as_min):
        as_def = as_min
    return(kmd_, kx_, kz_, as_prov, as_min, as_def)
```

#### Momento positivos - Verificação fadiga

```python
# Formulas

## Momento de inercia
def find_i(bw, x, AS, alfa_e, d):
    return ((bw*(x**3)/3) + (AS*alfa_e)* ((d-x)**2))

## Tensao concreto
def tensao_concreto(m, I, y):
    return ((m/I) * y)

## Tensao aco
def tensao_aco(alfa_e, m, I, d, x):
    return ((m/I) * (d-x) * alfa_e) * 1000

## Tensao aco negativa
def tensao_aco_neg(alfa_e, m, I, c, x):
    return ((m/I) * (x-c) * alfa_e) * -1000

def find_x(bw,alfa_e,AS,d):
    # a*x^2 + b*x + c = 0
    a = (bw/2)
    b = alfa_e * AS
    c = -alfa_e * AS * d
    x1, x2 = np.roots([a,b,c])
    if(x1 > x2):
#         print(x1)
        return x1
    else:
#         print(x2)
        return x2
```

```python
# coeficiente de transformacao aco -> concreto (nbr 6118:2014, 23.5.3)
alfa_e = 10

# coeficiente de reducao para TABULEIRO
psi_1 = 0.8
        
#Md de servico = mg + mq(amp) * psi_1
md_ser_pos = mg + np.multiply(mp_pos_amp, psi_1)

# Tabela momentos e geometria
df = pd.DataFrame()
df['Secao'] = secoes
df['x (m)'] = coordenadas_secoes
df['Mg'] = mg
df['Mq pos'] = mp_pos_amp
df['MD serv'] = md_ser_pos
df['bw (cm)'] = bw
df['d (cm)'] = d
df['h (cm)'] = h
df
# aplicar formula 'md_to_as' que retorna kmd, kx, kz, area de aco prov, area minima, e area definitiva
# a formula tem como parametro o MD, bw, d, h, fcd, fs
df2= pd.DataFrame(columns=('kmd', 'kx', 'kz', 'as_prov','as_min', 'as_def'))
for i in range(0, len(d)):
    # USAR MD_POS E NAO MD_SER_POS??????
    df2.loc[i] = (md_to_as(md_pos[i],bw,d[i],h[i],fcd,fs))

# concatenar a tabela anterior com o resultado obtido    
df_pos_ser = pd.concat([df,df2], axis=1)

# Achar X
x_pos = []
for i in range(0, len(md_ser_pos)):
    x_pos.append(find_x(bw, alfa_e, df_pos['as_def'][i], d[i]))
    
# Achar momento de inercia 2
i_pos = []
for i in range(0, len(md_ser_pos)):
    i_pos.append(find_i(bw, x_pos[i], df_pos_ser['as_def'][i],alfa_e, d[i]))
    
# Achar tensao armadura positiva
## tensao inf
tensao_as_inf = []
for i in range(0, len(md_ser_pos)):
    tensao_as_inf.append(tensao_aco(alfa_e, md_ser_pos[i],i_pos[i],d[i], x_pos[i]))

# Achar tensao armadura negativa
tensao_as_sup = []
for i in range(0, len(md_ser_pos)):
    tensao_as_sup.append(tensao_aco_neg(alfa_e, md_ser_pos[i],i_pos[i],3, x_pos[i]))

df_pos_ser['x'] = x_pos
df_pos_ser['I'] = i_pos
df_pos_ser['Tensao as inf mom_pos'] = tensao_as_inf
df_pos_ser['Tensao as sup mom_pos'] = tensao_as_sup

df_pos_ser[:]
```

#### Momentos negativos - Verificação fadiga

```python
#Md de servico = mg + mq(amp) * psi_1
md_ser_neg = mg + np.multiply(mp_neg_amp, psi_1)

# Tabela momentos e geometria
df = pd.DataFrame()
df['Secao'] = secoes
df['x (m)'] = coordenadas_secoes
df['Mg'] = mg
df['Mq pos'] = mp_neg_amp
df['MD serv'] = md_ser_neg
df['bw (cm)'] = bw
df['d (cm)'] = d
df['h (cm)'] = h
df
# aplicar formula 'md_to_as' que retorna kmd, kx, kz, area de aco prov, area minima, e area definitiva
# a formula tem como parametro o MD, bw, d, h, fcd, fs
df2= pd.DataFrame(columns=('kmd', 'kx', 'kz', 'as_prov','as_min', 'as_def'))
for i in range(0, len(d)):
    # USAR MD_POS E NAO MD_SER_neg??????
    df2.loc[i] = (md_to_as(md_neg[i],bw,d[i],h[i],fcd,fs))

# concatenar a tabela anterior com o resultado obtido    
df_neg_ser = pd.concat([df,df2], axis=1)

# Achar X
x_pos = []
for i in range(0, len(md_ser_neg)):
    x_pos.append(find_x(bw, alfa_e, df_neg['as_def'][i], d[i]))
    
# Achar momento de inercia 2
i_pos = []
for i in range(0, len(md_ser_neg)):
    i_pos.append(find_i(bw, x_pos[i], df_neg_ser['as_def'][i],alfa_e, d[i]))
    
# Achar tensao armadura positiva
## tensao inf
tensao_as_inf = []
for i in range(0, len(md_ser_pos)):
    tensao_as_inf.append(tensao_aco(alfa_e, -md_ser_neg[i],i_pos[i],d[i], x_pos[i]))

## tensao mg (apenas carga permanente)
tensao_as_mg = []
for i in range(0, len(md_ser_pos)):
    result = tensao_aco(alfa_e, -mg[i],i_pos[i],d[i], x_pos[i])
    # condicao para guardar apenas resultados acima de 0
    if(result >= 0):
        tensao_as_mg.append(result)
    else:
        tensao_as_mg.append(0)

# Achar tensao armadura negativa
tensao_as_sup = []
for i in range(0, len(md_ser_pos)):
    tensao_as_sup.append(tensao_aco_neg(alfa_e, -md_ser_neg[i],i_pos[i],3, x_pos[i]))

df_neg_ser['x'] = x_pos
df_neg_ser['I'] = i_pos
df_neg_ser['Tensao as inf mom_neg'] = tensao_as_inf
df_neg_ser['Tensao as mg'] = tensao_as_mg
df_neg_ser['Tensao as sup mom_neg'] = tensao_as_sup

df_neg_ser[:]
```

#### Resumo armadura positiva

```python
# Tabela momentos e geometria
df = pd.DataFrame()
df['Secao'] = secoes
df['x (m)'] = coordenadas_secoes
df_resumo_neg = pd.concat([
    df,
    df_pos_ser['Tensao as inf mom_pos'],
    df_neg_ser['Tensao as sup mom_neg'],
], axis=1)

# Variacao da tensao na armadura

### QUE VALOR EH ESSE ONDE ARMADURA NEGATTIVA 
delta = np.subtract(df_pos_ser['Tensao as inf mom_pos'],df_neg_ser['Tensao as sup mom_neg'])


# k fad
## considerando 2 * 10**6 (2 milhoes) de ciclos
delta_admis = 19 # kN/cm2 -- > 190 MPa
k_fad = []
for i in range(0, len(delta)):
    k_fad.append(delta[i]/delta_admis)

# k * As
k_as = np.multiply(df_pos_ser['as_def'], k_fad)

# Espacamento Armadura adotada
def espacamento(bitola, AS):
    area = np.pi*bitola**2/4
    provisorio = (100 / (AS/area))
    adotado = np.floor(provisorio)
    if(provisorio > 20):
        adotado = 20
    return (provisorio, adotado)

esp_prov = []
esp_adotado = []
for i in range(0, len(delta)):
    esp_prov.append(espacamento(1.25, k_as[i])[0])
    esp_adotado.append(espacamento(1.25, k_as[i])[1])

df_resumo_neg['Variacao tensao as'] = delta
df_resumo_neg['k fad'] = k_fad
df_resumo_neg['As'] = df_pos_ser['as_def']
df_resumo_neg['k * As'] = k_as
df_resumo_neg['esp prov'] = esp_prov
df_resumo_neg['esp adotado'] = esp_adotado
df_resumo_neg[:]
```
